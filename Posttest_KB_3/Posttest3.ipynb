## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv')
df
```

## Pie Chart

```python
data_pie = df['gender'].value_counts().rename_axis('gender').reset_index(name='Id_count')
plt.figure(figsize=(10, 12))
colors = sns.color_palette('Set2', 8)
plt.pie(data_pie.Id_count, labels=data_pie.gender, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Gender')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

dari grafik pie di atas kita dapat mengetahui bahwa wanita lebih banyak menderita Stroke

## Grafik Batang

```python
fig = plt.figure(figsize = (15,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Pekerjaan Penderita Stroke menurut Umur")
ax.set_xlabel("Umur")
ax.set_ylabel("Tipe Pekerjaan")
ax.bar(df["age"],df["work_type"], color = 'red')
plt.show(10)
```

dari grafik batang di atas kita dapat melihat untuk usia 0-20 adalah anak anak atau orang yang tidak pernah bekerja, sedangkan untuk usia 21 tahun ke atas rata rata merupakan pekerja di pemerintahan

```python
plt.figure(figsize=(10,5))
sns.heatmap(
    df.corr(),
    annot=True,
    cmap='flare'
) 
plt.title('korelasi antar attribut variabel ')

plt.show()
```
